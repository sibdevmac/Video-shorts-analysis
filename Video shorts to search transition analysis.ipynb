```python
platform_stats = df.groupby("platform")[["views","likes","comments","shares","saves"]].mean()
print(platform_stats)
```

```python
import matplotlib.pyplot as plt

platform_stats["views"].plot(kind="bar")
plt.title("Average Views by Platform")
plt.ylabel("Views")
plt.show()
```

```python
engagement = df.groupby("platform")[[
    "engagement_velocity",
    "engagement_total"
]].mean()
engagement.plot(kind="bar", figsize=(8,5))
plt.title("Engagement Comparison")
plt.show()
```

```python
top_hashtags = df["hashtag"].value_counts().head(20)
top_hashtags.plot(kind="bar")
plt.title("Top Discovery Topics")
plt.show()
```

```python
traffic = df["traffic_source"].value_counts()
traffic.plot(kind="bar")
plt.title("Top Discovery Topics")
plt.show()
```

```python
watch_stats = df.groupby("platform")[[
    "avg_watch_time_sec",
    "completion_rate"
]].mean()
watch_stats.plot(kind="bar")
plt.title("Watch Behavior Comparison")
plt.show()
```

```python
trend_duration = df.groupby("platform")["trend_duration_days"].mean()
print(trend_duration)
```

```python
creator_views = df.groupby("creator_tier")["views"].mean()
creator_views.plot(kind="bar")
plt.title("Views by Creator Tier")
plt.show()
```

```python
from collections import Counter
keywords = " ".join(df["title_keywords"].dropna())
word_counts = Counter(keywords.split())
word_counts.most_common(20)
```

```python
df["search_replacement_score"] = (
    df["engagement_rate"] +
    df["completion_rate"] +
    df["save_rate"]
)
score = df.groupby("platform")["search_replacement_score"].mean()
score
```

```python
from scipy.stats import ttest_ind

tiktok = df[df["platform"]=="TikTok"]["search_replacement_score"]
youtube = df[df["platform"]=="YouTube"]["search_replacement_score"]

t_stat, p_value = ttest_ind(tiktok, youtube)

print("T-stat:", t_stat)
print("P-value:", p_value)
```

The t-test result (T=150.36, p<0.001) indicates a statistically significant difference in engagement behavior between TikTok and YouTube Shorts, suggesting that short-form video platforms strongly influence how users discover information compared to traditional search mechanisms
